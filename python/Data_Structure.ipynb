# Data Structure

A data structure is a way of **organizing** and **storing data** so that it can be accessed and modified efficiently. Think of it as a container that holds data in a specific format, optimized for certain operations like searching, inserting, deleting, or sorting.


## Purpose
Makes data easier to use, process, and manage.

## Types
There are two main categories:

### 1. Primitive
Basic types like:
- Integers
- Floats
- Characters

### 2. Non-Primitive / Abstract Data Structures
Complex structures like:
- Arrays
- Linked Lists
- Stacks
- Queues
- Trees
- Graphs
- Hash Tables

## Operations
Depending on the structure, you can perform operations such as:
- Insert
- Delete
- Search
- Traverse
- Sort

## Data Structure Examples

### Array
- Stores elements in contiguous memory locations.
- Easy to access by index.
- Costly to insert or delete elements in the middle.

### Linked List
- Stores elements as nodes linked via pointers.
- Easy to insert and delete elements anywhere.
- Slower to access specific elements compared to arrays.

# 1️⃣ Linear Data Structures


These store elements sequentially.


## 1. Array (List in Python)
A collection of elements stored in contiguous memory locations.
**Key Properties**
1. Indexed access → O(1)
2. Fixed size (in low-level languages)
3. Python list is dynamic


##  Array: Time Complexity & Use Cases

### Time Complexity

| Operation        | Complexity       |
|-----------------|----------------|
| Access           | O(1)           |
| Insert (end)     | O(1) amortized |
| Insert (middle)  | O(n)           |
| Delete           | O(n)           |

### Use Cases
- Storing ordered data
- Lookup by index
- Base structure for many algorithms

# Types of Arrays

## 1. One-Dimensional Array
- Stores elements in a single linear sequence.
- Accessed using a single index.
- Example: `arr[5]`

## 2. Two-Dimensional Array
- Stores data in a table-like structure (rows and columns).
- Accessed using two indices: `arr[row][col]`.
- Example: `matrix[3][4]`

## 3. Multi-Dimensional Array
- Extends beyond 2 dimensions (3D, 4D, …).
- Used in complex data representations like matrices, tensors.
- Example: `tensor[2][3][4]`

## 4. Dynamic Array
- Size can grow or shrink during runtime.
- Example: Python `list`, Java `ArrayList`.

## 5. Jagged Array
- Each row can have different number of columns.
- Example: `arr = [[1,2,3], [4,5], [6]]` (Python)
