### PEP 8 Cheat Sheet (Quick Revision)
 ### 1. Indentation
- Use 4 spaces per indentation level
- Do not mix tabs and spaces

Example

```python
def hello():
   print("Hello")
```

#### 2. Line Length

- Maximum line length: 79 characters
- Comments/docstrings: 72 characters

 #### 3. Blank Lines

| Situation                              | Rule           |
|----------------------------------------|---------------|
| Between top-level functions/classes   | 2 blank lines |
| Between class methods                 | 1 blank line  |

 ####  4. Imports

**Rules:**
- One import per line
- Group imports in order:
- Standard library
- Third-party packages
- Local modules

#### 5. Naming Conventions

| Element    | Style       | Example           |
|------------|------------|------------------|
| Variables  | snake_case | user_age         |
| Functions  | snake_case | calculate_total()|
| Classes    | PascalCase | StudentRecord    |
| Constants  | UPPER_CASE | MAX_SIZE         |
| Modules    | snake_case | data_loader.py   |

#### 6. Spaces Around Operators

Always add spaces.

```python
# Good
x = a + b
total = price * quantity

# Bad
x=a+b
```

### 7. Spaces After Commas
` numbers = [1, 2, 3]`

### 8. Comparisons to None

Use is or is not

Good
- if value is None:

Bad
- if value == None:

### 9. Comments

- Start with #

- Keep comments clear and meaningful.

### 10. Docstrings

Use triple quotes.


```python
def add(a, b):
    """Return the sum of two numbers."""
    return a + b
```

#### 11. Avoid Extra Spaces

Good ✅

- my_list = [1, 2, 3]

Bad ❌

- my_list = [ 1, 2, 3 ]
