## Encapsulation

**Definition:**
-  Encapsulation is the practice of hiding the internal details of an object and exposing only what is necessary. 
- It bundles data (attributes) and methods (functions) that operate on that data into a single unit, i.e., a class.

**Purpose:**

- Protects data from unintended access or modification.

- Provides a clear interface to interact with the object.

- Improves maintainability and reduces complexity.

How it's done in Python:

- Public attributes/methods: Accessible from anywhere.

- Private attributes/methods: Use a single _ (protected) or double __ (private) prefix.

````python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance  # public method to access private data

# Usage
account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())  # 1500
# print(account.__balance)  # Error: cannot access private attribute
````

### Getter and Setter

**Definition:**
1. **Getter:** A method used to access (read) the value of a private attribute.
2. **Setter:** A method used to modify (write/update) the value of a private attribute, usually with validation.

**Purpose:**
- Control access to private attributes.
- Add validation when updating data.
- Maintain encapsulation.

````python
class Person:
    def __init__(self, name, age):
        self.__name = name   # private attribute
        self.__age = age     # private attribute

    # Getter for name
    def get_name(self):
        return self.__name

    # Setter for name
    def set_name(self, new_name):
        self.__name = new_name

    # Getter for age
    def get_age(self):
        return self.__age

    # Setter for age with validation
    def set_age(self, new_age):
        if new_age > 0:
            self.__age = new_age
        else:
            print("Invalid age!")

# Usage
person = Person("Bismah", 22)
print(person.get_name())  # Bismah
person.set_age(25)
print(person.get_age())   # 25
person.set_age(-5)        # Invalid age!
````

#### **@property:**
@property lets you access private attributes like normal attributes while still controlling reading/writing.

````python
class Person:
    def __init__(self, age):
        self.__age = age

    # Getter
    @property
    def age(self):
        return self.__age

    # Setter
    @age.setter
    def age(self, value):
        if value > 0:
            self.__age = value
        else:
            print("Invalid age!")

# Usage
p = Person(22)
print(p.age)   # 22 (uses getter)
p.age = 30     # (uses setter)
print(p.age)   # 30
p.age = -10    # Invalid age!
````

## Polymorphism in OOP
The same function or method name can work with different types of objects and perform different behavior.

#### **Types of Polymorphism**
Polymorphism is mainly classified into two types:
1.  Compile-Time (Static) Polymorphism / method overloading ()
2.  Run-Time (Dynamic) Polymorphism  / method overriding  (same method name in praent and child but behave differntly)
3.  Operator Polymorphism (behave different base on datatype) 
4. Function Polymorphism  ( function behave differently base on datatype)

1. Compile-Time (Static) Polymorphism

    - Resolved during compilation.

      Achieved using:
      1. Method Overloading (same method name, different parameters)
      2. Operator Overloading (same operator behaves differently for different types)



**Example : Method Overloading (Python workaround)**

````python

class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

calc = Calculator()
print(calc.add(5, 10))    # Output: 15
print(calc.add(5, 10, 20)) # Output: 35
````

#### Run-Time (Dynamic) Polymorphism

- Resolved during execution.

Achieved using:

1. Method Overriding (subclass provides its own version of a superclass method)
2. Duck Typing (Python specific: “if it walks like a duck and quacks like a duck, it’s a duck”)

**Example: Method Overriding**

````python
class Animal:
    def sound(self):
        print("Some generic sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

animals = [Dog(), Cat(), Animal()]

for a in animals:
    a.sound()
````

#### **Advantages of Polymorphism**

- Code Reusability: Same interface works for multiple types.

- Flexibility & Maintainability: Adding new classes does not require changing existing code.

- Extensibility: Easy to extend programs without modifying existing code.

- Simplifies Interface: Users interact with a single method name rather than multiple methods for each class.

## **Abstraction**
Abstraction is the process of hiding the internal implementation details of a class or method and showing only the essential features to the user. It allows the programmer to focus on what an object does, rather than how it does it.

**Key Points**

1. **Focus on essentials:** Only expose the necessary methods or attributes.
2. **Hides complexity:** Internal code is hidden from the user, reducing errors and improving usability.
3. **Supports modularity:** Changes in internal implementation don’t affect code using the abstraction.
4. **Implemented via:**
     - Abstract classes
     - Interfaces (in languages that support them)

> Python provides abstraction using abstract classes in the abc module.

````python
from abc import ABC, abstractmethod

# Abstract class
class Vehicle(ABC):

    @abstractmethod
    def start_engine(self):
        pass  # Abstract method, no implementation

# Concrete class
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started!")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine started!")

# Using the abstraction
vehicles = [Car(), Bike()]
for v in vehicles:
    v.start_engine()
````

**Benefits of Abstraction**

- **Simplifies usage:** Users don’t need to understand inner workings.
- **Reduces complexity:** Hides complicated logic.
- **Enhances security:** Sensitive data and methods can be hidden.
- **Supports code maintenance:** Internal changes don’t affect external code.
- **Promotes reusability:** Abstract methods can be implemented differently in multiple subclasses.

**Components of Abstraction**
1. Abstract Method
   - Abstract methods are method declarations without a body defined inside an abstract class.
   - They act as placeholders that force subclasses to provide their own specific implementation, ensuring consistent structure across derived classes.
2. Concrete Method
   - Concrete methods are fully implemented methods within an abstract class.
   - ubclasses can inherit and use them directly, promoting code reuse without needing to redefine common functionality.
3. Abstract Properties
   - Abstract properties work like abstract methods but are used for properties
   -  These properties are declared with @property decorator and marked as abstract using @abstractmethod.
   - Subclasses must implement these properties.
4. Abstract Class Instantiation
   - Abstract classes cannot be instantiated directly
   - This is because they contain one or more abstract methods or properties that lack implementations. 
   - Attempting to instantiate an abstract class results in a TypeError.

## **Inheritance**
Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a class (called a child or derived class) to inherit attributes and methods from another class (called a parent or base class).

#### **Key Concepts**
1. Parent (Superclass)
   - The class whose properties and methods are inherited.
   - Example: Vehicle class with attributes like speed and color.

2. Child (Subclass)
    - The class that inherits from the parent.
    - Can use, override, or extend the parent’s properties and methods.
    - Example: Car class inherits from Vehicle and adds number_of_doors

#### **Types of Inheritance**

1. Single Inheritance
    - Child inherits from one parent class.
2. Multiple Inheritance
    - Child inherits from more than one parent class.
3. Multilevel Inheritance
    - A chain of inheritance (grandparent → parent → child).
4. Hierarchical Inheritance
    - Multiple children inherit from the same parent class.
5. Hybrid Inheritance
    - Combination of two or more inheritance types.


#### **super() function**
- The super() function is used in a child class to call methods or access properties of its parent class.
- It is especially useful when overriding methods but still needing the parent’s behavior.
> super().method_name(arguments)
   - super() returns a temporary object of the parent class.
   - .method_name() calls the method from the parent.

#### **Diamond Problem**
- The Diamond Problem occurs in multiple inheritance when a class inherits from two classes that both inherit from the same parent class.
- This creates ambiguity about which path to follow to access the parent class method.

#### **Python’s Solution: MRO (Method Resolution Order)**
- Python solves the diamond problem using C3 Linearization (MRO), which defines a consistent order to look for methods.

**MRO**
MRO (Method Resolution Order) is the order in which Python looks for a method or attribute in a hierarchy of classes.
**How it work**
- Python uses C3 linearization to determine the search order.

#### Composition vs Aggregation

| Concept     | Description                                                                                       | Example                                           |
|------------|---------------------------------------------------------------------------------------------------|--------------------------------------------------|
| Composition | A strong “has-a” relationship where the child object cannot exist without the parent object. If the parent is destroyed, the child is destroyed too. | A Car has an Engine. The engine cannot exist independently outside the car. |
| Aggregation | A weaker “has-a” relationship where the child object can exist independently of the parent.      | A Teacher and a Department. A teacher can exist without being in a department. |

## Key Differences: Composition vs Aggregation

| Feature             | Composition                  | Aggregation                     |
|--------------------|------------------------------|---------------------------------|
| Relationship Strength | Strong                       | Weak                            |
| Lifetime Dependency   | Child depends on parent      | Child independent of parent     |
| Ownership            | Parent owns the child        | Parent references the child     |
| Example              | Car → Engine                 | Department → Teacher            |

**a. Composition**

````python
class Engine:
    def __init__(self, power):
        self.power = power

class Car:
    def __init__(self, brand, power):
        self.brand = brand
        self.engine = Engine(power)  # Engine is part of Car (composition)

my_car = Car("Toyota", 120)
print(my_car.engine.power)  # Output: 120
````

**b. Aggregation**

````python
class Teacher:
    def __init__(self, name):
        self.name = name

class Department:
    def __init__(self, dept_name, teacher):
        self.dept_name = dept_name
        self.teacher = teacher  # Reference to Teacher (aggregation)

t1 = Teacher("Bismah")
dept = Department("CS", t1)
print(dept.teacher.name)  # Output: Bismah
````
