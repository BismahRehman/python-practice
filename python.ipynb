# What is Python?

Python is one of the most popular programming languages. It’s simple to use, packed with features and supported by a wide range of libraries and frameworks. Its clean syntax makes it beginner-friendly

- A high-level language, used in data science, automation, AI, web development and more.
- Known for its readability, which means code is easier to write, understand and maintain.
- Backed by strong library support, we don’t have to build everything from scratch

## Python Excution Workflow
- Python uses a compiler when converting .py → bytecode.

- Python uses an interpreter when executing the bytecode.

```python


Python Source Code (.py)
        ↓
Compilation
        ↓
Bytecode (.pyc)
        ↓
Python Virtual Machine
        ↓
Program Output
```

### How Python REPL Works

The cycle looks like this:

> Read → Evaluate → Print → Loop

- Read – Python reads the command you type

- Evaluate – Python executes the command

- Print – Python shows the result

- Loop – Python waits for the next command

 **Repl is used by pyhton developer on Command Line**
 ### Why Developers Use REPL

REPL is useful for:

- Quick testing of code

- Learning Python

- Debugging small problems

- Trying library functions
